Downlaod solar data from RE ninja

```python
## produce solar data from RE ninja; based on location data

year = 2020

df_data_solar = pd.DataFrame()

df_location = pd.read_excel('input\\Taiwan RE location.xlsx')
df_location.fillna('', inplace=True)
df_location_pv = df_location.loc[ df_location['類別'] == '太陽光電']
df_location_pv.reset_index(drop=True, inplace=True)
df_location_pv
```

```python
## produce solar data from RE ninja; based on location data
## NOTE: RE ninja can only run about 6 request in a certain period of time; need to wait a few minute before next run
n = 0
for i in range(n,n+6):

    rowdata = df_location_pv.loc[i]
    df_REninja_pv = REninja_pv('pv', lat=rowdata['緯度'], lon=rowdata['經度'], year=year)

    if '時間' not in df_data_solar.columns:
        df_data_solar['時間'] = df_REninja_pv['local_time']

    col_cf = '光電-容量因素-' + rowdata['地點-縣市']
    df_data_solar[ col_cf ] = df_REninja_pv['electricity']

    col_power = '光電-發電量-' + rowdata['地點-縣市']
    df_data_solar[ col_power ] = df_REninja_pv['electricity'] * rowdata['總容置容量 (MW)']

df_data_solar
```

Download wind data from RE ninja

```python
## produce wind data from RE ninja; based on location data
## load location data
import pandas as pd
from functions import *

df_location = pd.read_excel('input\\Taiwan RE location.xlsx')
df_location.fillna('', inplace=True)

year = 2020

df_data_wind = pd.DataFrame()

df_location_wind = df_location.loc[ df_location['類別'] == '陸域風電' ]
df_location_wind = pd.concat([df_location_wind, df_location.loc[ df_location['類別'] == '離岸風電' ] ])
df_location_wind.reset_index(drop=True, inplace=True)
df_location_wind
```

```python
## produce wind data from RE ninja; based on location data
## NOTE: RE ninja can only run about 6 request in a certain period of time; need to wait a few minute before next run

n = 0
for i in range(n,n+5):

    rowdata = df_location_wind.loc[i]
    df_REninja_pv = REninja_pv('wind', lat=rowdata['緯度'], lon=rowdata['經度'], year=year)

    if '時間' not in df_data_wind.columns:  df_data_wind['時間'] = df_REninja_pv['local_time']

    col_cf = rowdata['類別'] + '-容量因素-' + rowdata['地點-縣市']
    df_data_wind[ col_cf ] = df_REninja_pv['electricity']

    col_power = rowdata['類別'] + '-發電量-' + rowdata['地點-縣市']
    df_data_wind[ col_power ] = df_REninja_pv['electricity'] * rowdata['總容置容量 (MW)']

df_data_wind
```

```python
## produce wind data from RE ninja; based on location data
## NOTE: RE ninja can only run about 6 request in a certain period of time; need to wait a few minute before next run

n = 7
for i in range(n,n+3):

    rowdata = df_location_wind.loc[i]
    df_REninja_pv = REninja_pv('wind', lat=rowdata['緯度'], lon=rowdata['經度'], year=year)

    if '時間' not in df_data_wind.columns:  df_data_wind['時間'] = df_REninja_pv['local_time']

    col_cf = rowdata['類別'] + '-容量因素-' + rowdata['地點-縣市']
    df_data_wind[ col_cf ] = df_REninja_pv['electricity']

    col_power = rowdata['類別'] + '-發電量-' + rowdata['地點-縣市']
    df_data_wind[ col_power ] = df_REninja_pv['electricity'] * rowdata['總容置容量 (MW)']

df_data_wind
```
